## `transform`

```python
from torchvision.transforms import ToTensor, Compose

t = Compose([ ToTensor() ])
```

```python
dataset = torchvision.datasets.MNIST(
    root="datasets/", 
    download=False, 
    train=True,
    transform=t
)
```

```python
dataset[100][0].shape
```
